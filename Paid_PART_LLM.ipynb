```python
#@title 4. ChatPromptTemplate & chain

from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.prompts.chat import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
)                                                                                                            

chat = ChatOpenAI(model_name='gpt-4',temperature=0)

# 번역하는거 해라하는 프롬프트
template='''
너는 개인정보처리방침을 진단하는 솔루션이야. 
{policy}는 이 회사의 방침이야 그리고 {instruction}은 개인정보처리 지침이야.
개인정보처리 지침을 보고 이 회사의 방침에서 잘못된 부분이 있다면 찾아서 설명해줘!
'''

#시스템 메시지로 
system_message_prompt = SystemMessagePromptTemplate.from_template(template)

# 휴먼 메시지 추가 가능!
chat_prompt = ChatPromptTemplate.from_messages([system_message_prompt])

chatchain = LLMChain(llm=chat, prompt=chat_prompt)
```

# 2) 대제목 커팅하는 알고리즘
## 1> {'방침 대제목'  : index}로 정렬된 딕셔너리 추출
## 2> 데이터 프레임의 rule에 매칭되는 행에 파트를 추가

```python
part_dict = {}
part_dict = {key: '' for key in title_dict}


for key in part_dict:
    index = bob.find(key)
    part_dict[key] = index
part_dict = dict(sorted(part_dict.items(), key=lambda item: item[1]))
```

```python
part_dict
```

```python
result_dict = {key: part_dict[value] for key, value in title_dict2.items()}
```

```python
result_dict
```

```python
keys = list(result_dict.keys())
for i in range(len(keys)):
    # 마지막 키인 경우 텍스트 끝까지 자른다.
    if i == len(keys) - 1:
        result_dict[keys[i]] = bob[result_dict[keys[i]]:]
    else:
        # 그 외의 경우 다음 인덱스까지 자른다.
        result_dict[keys[i]] = bob[result_dict[keys[i]]:result_dict[keys[i + 1]]]
```
